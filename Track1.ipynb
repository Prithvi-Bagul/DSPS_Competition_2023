## **Creation Of Datset**
For training model, datsets were created by annotation of datasets with the different 07 classes. And this annotated images were split in the 80 and 20 percent of training and validation dataset, which is stored in Yolo_data in training and validation dataset

```python
import os
from simple_fxn import cvat2yolo
# location of cvat annotations
if not os.path.isdir('training_data/'):
  print ('Training Data Missing')

if not os.path.isdir('yolo_data/'):
  os.makedirs('yolo_data')
annotation_folder = 'training_data/'
output_path = 'yolo_data/training'
val_path = 'yolo_data/validation'
for annts in os.listdir(annotation_folder):
  json_file = os.path.join(annotation_folder,annts,'annotations','instances_default.json')
  img_path = os.path.join(annotation_folder,annts,'images')
  if(os.path.isfile(json_file)):
    cvat2yolo(annts,json_file,img_path, output_path, val_path)
```

## **APPROACH FOLLOWED FOR TRAINING**

For training the model, we used 80 and 20 ratio as training and validation images. For training model, we used 165 epochs and batch-size as 16. Also, install the alubmentations with the help of "!pip install -U albumentations
!pip install opencv-python-headless<4.3" 
for augementation purpose, in utils folder, there is augmentations.py file is there, so using text editor, following values were modified. 
*   A.RandomSizedBBoxSafeCrop(p=0.5, height=640, width=640, erosion_rate=0.2),
*   A.InvertImg(p=0.5),

For training model, we used same Yolov5 model and we used Adam optimizer instead of SGD as it is most suitable for multiple classes. So, we change the learning and momentum rate and other values which are listes. So, hyperparameter were changed using hyp.scratch.high.yaml which is in the data/hyps folder. 
*   lr0: 0.001
*   momentum: 0.999
*   flipud: 0.5
*   box: 0.01
*   cls: 0.1
These are the above values are changed to prevent overfitting of model. So, we get the the results are saved in runs/train/exp13 folder.
Actually the results are not shown of results of exp13 is not shown, becaue we also tried to run the model for improving accuracy, results of exp13 is shown in Track1_results.csv, where we tried to run for 200 epochs, but it was stopped at 165 iteration only.
